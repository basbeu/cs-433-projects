```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

val_800 = pd.read_csv('run-20191206-160757_validation-tag-epoch_accuracy.csv')
val_10000 = pd.read_csv('run-20191206-233429_validation-tag-epoch_accuracy.csv')

plt.figure(figsize=(6,4))
val_800.Value[0:400].plot(label='800 training images',color='red')
ax = val_10000.Value.plot(label="10'000 training images",color='green')
ax.legend(loc=4)
ax.set_ylim([0.92,1.0])
ax.set_xticks(np.arange(9)*50)
ax.set_xlabel('Number of epochs')
ax.set_ylabel('Validation set accuracy')
plt.savefig('validation_graph.png', dpi = 300);
```
